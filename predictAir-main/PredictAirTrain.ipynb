```python
  import tensorflow as tf
  from tensorflow import keras
  import pandas as pd
  import numpy as np
  import pickle
```

```python
df = pd.read_excel('/content/Data.xlsx')
```

```python
df
```

```python
label_column = df[df.columns[[-1]]]
df = df.drop(df.columns[[-1]],axis=1)
```

```python
label_column
```

```python
x = df.to_numpy()
y = label_column.to_numpy()
```

```python
# Split the data into training and test sets
train_ratio = 0.8  # 80% for training, 20% for testing
train_samples = int(len(x) * train_ratio)

X_train = x[:train_samples]
y_train = y[:train_samples]
X_test = x[train_samples:]
y_test = y[train_samples:]

# Create tf.data.Dataset objects for training and testing
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train))
test_dataset = tf.data.Dataset.from_tensor_slices((X_test, y_test))

# Shuffle and batch the training dataset
train_dataset = train_dataset.shuffle(buffer_size=len(X_train)).batch(batch_size=8)

# Batch the test dataset
test_dataset = test_dataset.batch(batch_size=8)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1,activation='linear'),
])
```

```python
model.compile(optimizer='adam',
              loss='mean_squared_error'
)
```

```python
model.fit(train_dataset,epochs=250)
```

```python
# Evaluate the model on the test dataset
loss = model.evaluate(test_dataset)
print(f"Test set loss: {loss}")
```

```python
model.predict([[15,0.9,0.9,0.9]])
```
